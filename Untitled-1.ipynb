```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# Inicializa el controlador de Selenium (necesitas el controlador de tu navegador)
driver = webdriver.Chrome()  # Cambia a tu navegador preferido
```

```python
# Abre la página web con el gráfico
url = "https://worldmigrationreport.iom.int/wmr-2020-interactive/?lang=ES"
driver.get(url)
```

```python
# Obtén el contenido HTML de la página
html_content = driver.page_source
```

```python
# Cierra el controlador de Selenium
driver.quit()
```

```python
# Crea un objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# Encuentra el elemento que contiene el gráfico
chart_element = soup.find("div", class_="lineChart__chart.svelte-kbowc9")
```

```python
# Encuentra todos los elementos <path> dentro del elemento del gráfico
path_elements = chart_element.find_all("path")
```

```python
# Extrae y muestra los datos de los atributos "d" de los elementos <path>
for path in path_elements:
    data = path["d"]
    print("Datos:", data)
    print("----")
```
