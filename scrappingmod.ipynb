```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

wait = WebDriverWait(driver, 30)  # Aumenta el tiempo de espera si es necesario
symbols = wait.until(EC.visibility_of_element_located((By.CLASS_NAME, 'symbols')))

# Buscar elementos circle dentro del elemento symbols
circles = symbols.find_elements(By.TAG_NAME, 'circle')

# Hacer algo con los elementos circle (por ejemplo, imprimir sus atributos)
for circle in circles:
    print(circle.get_attribute('cx'), circle.get_attribute('cy'))
```

```python
from selenium.webdriver.common.action_chains import ActionChains

# Crear una instancia de ActionChains
action = ActionChains(driver)

# Encontrar todos los elementos <g class="hover__shapes">
hover_shapes = driver.find_elements(By.CLASS_NAME, 'hover__shapes')

data_list = []

for hover_shape in hover_shapes:
    # Realizar el hover sobre el elemento
    action.move_to_element(hover_shape).perform()

    time.sleep(5)

    # Esperar a que aparezca la caja de información
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'info-box')), 20)

    # Extraer los datos de la caja de información
    info_box = driver.find_element(By.CLASS_NAME, 'info-box')
    country_name = info_box.find_element(By.CLASS_NAME, 'country').text
    total_population = info_box.find_element(By.CLASS_NAME, 'total-population').text
    migrants_percentage = info_box.find_element(By.CLASS_NAME, 'migrants-percent').text
    non_migrants_percentage = info_box.find_element(By.CLASS_NAME, 'nonmigrants-percent').text

    data = {
        'País': country_name,
        'Total de habitantes': total_population,
        '% de migrantes': migrants_percentage,
        '% de no migrantes': non_migrants_percentage
    }

    data_list.append(data)

print(data_list)
```
